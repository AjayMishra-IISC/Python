```python
def select(n: int) -> bool:
    return n % 3 == 0 or n % 5 ==0

def euler001_1(limit: int) -> int:
    n, total = 1, 0
    while n < limit:
        if select(n):
            total += n
        n +=1
    return total
```

```python
def is_prime(n: int) ->int:
    for i in range(2, n):
        if n%i == 0:
            return False
    return True
```

```python
def euler002_01(limit: int) ->int:
    sum = count =0
    for i in range(2, int(limit+1)):
        if is_prime(i):
            print(i)
            sum = sum + i
            count += 1
    print ("Sum of Prime numbers till ", limit, "is = ",sum)
```

```python
def euler002_01(limit: int) ->int:
    primes = set()
    notPrimes = set()
    i = 2
    while len(primes)<limit//2:
        if not i in notPrimes:
            primes.add(i);
            for notPrime in range(i*2, limit, i):
                notPrimes.add(notPrime)
        i+=1
    print("Sum of number till ", limit, "is :", sum(primes))
```

```python
def Is_Divisible(Divisor: int, Divident: int) -> bool:
    i, rem = 3, 0
    while(i <= Divisor):
        rem = Divident % i
        i += 1
        if (rem != 0):
            return False
    return True
```

```python
def num_to_digits(n: int) -> list[int]:
    return [int(ch) for ch in str(n)]
```

```python
def digits_to_num(ds: list[int]) -> int:
    if len(ds) == 1:
        return ds[-1]
    elif len(ds) > 1:
        return ds[-1] + 10 * digits_to_num(ds[:-1])
```

```python
def largest(n: int) -> int:
    return digits_to_num(sorted(num_to_digits(n), reverse=True))

def smallest(n: int) -> int:
    return digits_to_num(sorted(num_to_digits(n)))
```

```python
def next_kap(n: int) -> int:
    return largest(n) - smallest(n)
```

```python
def euler007_01(n: int) -> list[int]:
    seq = []
    while n not in seq:
        seq.append(n)
        n = next_kap(n)
    return seq
```

```python
def multiplyList(myList) ->int:
    # Multiply elements one by one
    result = 1
    for x in myList:
        result = result * int(x)
    return result
```
