```python
def FizzBuzz (num: int) -> int:
    val = 0
    if (num % 3 == 0):
        val = 1
    if (num % 5 == 0):
        val += 2
    return val
```

```python
def euler005_01(limit: int) ->int:

    for n in range(1, limit+1):
        i = FizzBuzz(n)
        if (i == 0):
         print(i)
        if (i == 1):
         print("Fizz")
        if (i == 2):
         print("Buzz")
        if (i == 3):
         print("FizzBuzz")
```

```python
euler005_01(30)
```

```python
def Is_Divisible(Divisor: int, Divident: int) -> bool:
    i, rem = 3, 0
    while(i <= Divisor):
        rem = Divident % i
        i += 1
        if (rem != 0):
            return False
    return True
```

```python
def euler006_01(value: int) ->int:
    if (value < 3):
        print("Value :", value)
        return
    count = 2
    Flag = False
    while(count):
        count += 1
        Flag = Is_Divisible(value, count)
        if(Flag == True):
            break
    print("Smallest number divisible by ", value, "is :", count)
```

```python
euler006_01(20)
```

```python
def Is_Palindrome (maxNum: int) -> int:
    print("value Received: ", maxNum)
    loopCnt, iValue = 1, 0
    firstNum, secNum = maxNum, maxNum-1
    print("1st Num: ", firstNum, "2ndNum", secNum)

    while(loopCnt < maxNum/2):
        iValue = firstNum * secNum
        print(iValue)
        if((str(iValue)) == (str(iValue)[::-1])):
            print("end of loop")
            return iValue
        else:
            loopCnt +=1
            secNum  -=1
    return 0
```

```python
import numpy as np

def euler004_01(digit: int) -> int:
    print("Number of digits", digit)
    value = 9
    while(digit > 1):
        value += 9 * np.power(10, digit-1)
        digit -= 1
        print("Value: ", value, "Digit", digit)
    print("Value Passed", value)
    result = Is_Palindrome(value)
    num = value
    #result = 1
    while(result == 0 and num > value/2):
        result = Is_Palindrome(num)
        num -= 1
        print("num", num, "result", result)
    print("result is : ", result)
```

```python
def euler004_02(num: int) ->int:
    iPalindrome = []
    print(num)
    for i in range(num, 1, -1):
        for j in range(num, 1, -1):
            product = j * i
            if((str(product)) == (str(product)[::-1])):
                iPalindrome.append(product)
                print("Palindrome", product, "i", i,"j:", j)
    print("Highest Palindrome upto is : ", max(iPalindrome))
```

```python
def multiplyList(myList) ->int:
    # Multiply elements one by one
    result = 1
    for x in myList:
        result = result * int(x)
    return result
```
